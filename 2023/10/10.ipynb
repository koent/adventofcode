```python
from math import *

with open('input.txt', 'r') as f:
    grid = f.readlines()

visited = [[0 for p in line] for line in grid]
# 0 not touched, 1 left turn, 2 right turn, 3 straight


startpos = None
for y, line in enumerate(grid):
    if startpos != None:
        break
    for x, c in enumerate(line):
         if c == 'S':
             startpos = (x, y)
             break 

# I know we can start by going down
dir = (0, 1) # (dx, dy)
pos = (startpos[0], startpos[1] + 1) # already make a step
visited[startpos[1]][startpos[0]] = 3 # I know that S is on a |

nsteps = 1
while pos != startpos:
    c = grid[pos[1]][pos[0]]
    if c == 'L' and dir == (0, 1):
        visited[pos[1]][pos[0]] = 1
        dir = (1, 0)
    elif c == 'L' and dir == (-1, 0):
        visited[pos[1]][pos[0]] = 2
        dir = (0, -1)
    elif c == 'J' and dir == (0, 1):
        visited[pos[1]][pos[0]] = 2
        dir = (-1, 0)
    elif c == 'J' and dir == (1, 0):
        visited[pos[1]][pos[0]] = 1
        dir = (0, -1)
    elif c == '7' and dir == (1, 0):
        visited[pos[1]][pos[0]] = 2
        dir = (0, 1)
    elif c == '7' and dir == (0, -1):
        visited[pos[1]][pos[0]] = 1
        dir = (-1, 0)
    elif c == 'F' and dir == (0, -1):
        visited[pos[1]][pos[0]] = 2
        dir = (1, 0)
    elif c == 'F' and dir == (-1, 0):
        visited[pos[1]][pos[0]] = 1
        dir = (0, 1)
    else:
        visited[pos[1]][pos[0]] = 3

    nsteps += 1
    pos = (pos[0] + dir[0], pos[1] + dir[1])

## status:
# 5 outside
# 7 horizontal border
# 8 inner

n_inside = 0
for line in visited:
    status = 5
    for c in line:
        if status == 5:
            if c == 0:
                pass
            elif c == 1:
                status = 7
            elif c == 2:
                raise Exception("cant")
            elif c == 3:
                status = 8
        elif status == 7:
            if c == 0:
                raise Exception("cant")
            elif c == 1:
                status = 5
            elif c == 2:
                status = 8
            elif c == 3:
                pass
        elif status == 8:
            if c == 0:
                n_inside += 1
            elif c == 1:
                raise Exception("cant")
            elif c == 2:
                status = 7
            elif c == 3:
                status = 5

print(n_inside)
```
